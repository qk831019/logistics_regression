```python
import tables
import random

class Level(tables.IsDescription):
    name = tables.StringCol(16)
    index = tables.Int32Col()
    x = tables.Float32Col()
    y = tables.Float32Col()
    z = tables.Float64Col()


h5_file = tables.openFile('Level.h5', mode = 'w', title = 'title')
group = h5_file.createGroup('/', 'level', 'title2')
tab = h5_file.createTable(group, 'raw_data', Level, 'title3')

## Write with PyTables
level = tab.row

for i in range(10):
    level['name'] = 'value: %3d' % i
    level['index'] = i
    level['x'] = random.random()*10
    level['y'] = random.random()*10
    level['z'] = random.random()*10
    level.append()

tab.flush()

h5_file.close()
```
